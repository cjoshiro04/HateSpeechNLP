```python
def clean_and_preprocess(text):
    # 1. Convert text to lowercase
    text = text.lower()

    # 2. Remove special characters, numbers, and punctuation
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 3. Tokenize the text into words
    tokens = word_tokenize(text)

    # 4. Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # 5. Lemmatize words (convert words to their base form)
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # 6. Join the preprocessed tokens back into a clean text
    cleaned_text = ' '.join(tokens)

    return cleaned_text
```

```python
def lda_topic_vector(text, num_topics=5, passes=10):
    # Tokenize and preprocess the text
    tokens = simple_preprocess(text)
    if len(tokens) == 0:
        return False

    # Create a dictionary from the tokenized text
    dictionary = corpora.Dictionary([tokens])

    # Create a bag of words representation of the text
    bow_corpus = [dictionary.doc2bow(tokens)]

    # Build the LDA model
    lda_model = LdaModel(bow_corpus, num_topics=num_topics, id2word=dictionary, passes=passes)

    # Get the topic vector for the text
    topic_vector = lda_model[bow_corpus[0]]

    return topic_vector
```

```python
Hashtags = list(data["Hashtags"])
Captions = list(data["Captions"])
```

```python
cleaned_Hashtags_ = []
cleaned_Captions_ = []

similarity_score = []
for i in range(len(data)):
    cleaned_Hashtags = clean_and_preprocess(Hashtags[i])
    cleaned_Captions = clean_and_preprocess(Captions[i])

    cleaned_Hashtags_.append(cleaned_Hashtags)
    cleaned_Captions_.append(cleaned_Captions)

    Hashtags_vector = lda_topic_vector(cleaned_Hashtags, num_topics=5, passes=10)
    Captions_vector = lda_topic_vector(cleaned_Captions, num_topics=5, passes=10)

    if Hashtags_vector == False or Captions_vector == False:
        similarity_score.append("Invalid")
    else:
        similarity = gensim.matutils.cossim(Hashtags_vector, Captions_vector)
        similarity_score.append(similarity)
```

```python
similarity_score
```

```python
minimum = min(e for e in similarity_score if isinstance(e, float))
maximum = max(e for e in similarity_score if isinstance(e, float))
mid_val = (minimum + maximum)/2
mid_val
```

```python
count_invalid = 0
count_below = 0
count_above = 0

for ele in similarity_score:
    if type(ele) == str:
        count_invalid +=1
    else:
        if ele <= mid_val:
            count_below +=1
        else:
            count_above +=1
```

```python
print("Invalid : ", count_invalid)
print("Below Mid-value Threshold :", count_below)
print("Above Mid-value Threshold :", count_above)
```

```python
lables = list(data["Clickbait"])
```

```python
invalid_1 = 0
invalid_0 = 0

True_clickbait = 0
False_clickbait = 0

True_nonclickbait = 0
False_nonclickbiat = 0

for i in range(len(lables)):
    if type(similarity_score[i]) == str:
        if lables[i] == 1:
            invalid_1 += 1
        else:
            invalid_0 += 1

    else:
        if similarity_score[i] <= mid_val and lables[i] == 1:
            True_clickbait +=1
        elif similarity_score[i] <= mid_val and lables[i] == 0:
            False_clickbait +=1
        elif similarity_score[i] > mid_val and lables[i] == 0:
            True_nonclickbait +=1
        elif similarity_score[i] > mid_val and lables[i] == 1:
            False_nonclickbiat +=1
```

```python
accuracy = (True_clickbait + True_nonclickbait) / (True_clickbait + False_clickbait + True_nonclickbait + False_nonclickbiat)
accuracy
data
```

```python
data["LDA Feature"] = 0

for i in range(len(similarity_score)):
    if (similarity_score[i] == "Invalid"):
      data["LDA Feature"].iloc[i] = 0
    elif (similarity_score[i] > (float(minimum) + float(maximum))/2):
      data["LDA Feature"].iloc[i] = 1
    else:
      data["LDA Feature"].iloc[i] = 0

data = data.drop(columns=["SearchedTag", "Hashtags", "Captions"])
data
```

```python
#SVM Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import metrics

#Logistic Regression Imports
from sklearn import linear_model

#KNN Imports
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

def train_svm(df, df_xcols, df_ycol, kernel_type):
    x_svm = df[df_xcols]
    y_svm = df[df_ycol]

    X_train, X_test, y_train, y_test = train_test_split(x_svm, y_svm, test_size=0.3) # 70% training and 30% test

    clf = SVC(kernel=kernel_type)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    #Model Accuracy
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

    #Model Precision
    print("Precision:",metrics.precision_score(y_test, y_pred))

    #Model Recall
    print("Recall:",metrics.recall_score(y_test, y_pred))

    #Model F1 Score
    print("F1 Score: ", metrics.f1_score(y_test, y_pred))

    return clf

def train_logr(df, feature_cols, df_ycol):
    x_logr = df[feature_cols]
    y_logr = df[df_ycol]

    X_train, X_test, y_train, y_test = train_test_split(x_logr, y_logr, test_size=0.3)

    logr = linear_model.LogisticRegression(max_iter = 1000)

    logr.fit(X_train, y_train.values.ravel())
    y_pred = logr.predict(X_test)

    #Model Accuracy
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

    #Model Precision
    print("Precision:",metrics.precision_score(y_test, y_pred))

    #Model Recall
    print("Recall:",metrics.recall_score(y_test, y_pred))

    #Model F1 Score
    print("F1 Score: ", metrics.f1_score(y_test, y_pred))

    return logr

def train_KNN(df, feature_cols, df_ycol, neighbors):

    x_KNN = df[feature_cols]
    y_KNN = df[df_ycol]

    X_train, X_test, y_train, y_test = train_test_split(x_KNN, y_KNN, test_size = 0.3)

    knn = KNeighborsClassifier(n_neighbors=neighbors)

    knn.fit(X_train, y_train)

    y_pred = knn.predict(X_test)

    #Model Accuracy
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

    #Model Precision
    print("Precision:",metrics.precision_score(y_test, y_pred))

    #Model Recall
    print("Recall:",metrics.recall_score(y_test, y_pred))

    #Model F1 Score
    print("F1 Score: ", metrics.f1_score(y_test, y_pred))

    return knn
```
